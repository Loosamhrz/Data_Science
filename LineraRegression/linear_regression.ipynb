```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
data1 = pd.DataFrame({
    'X': [1400, 1600,1700, 1875, 1100, 1550, 2350, 2450, 1425, 1700],
    'Y': [245, 312, 279, 308, 199, 219, 405, 324, 319, 255]
})
data1
```

```python
sns.scatterplot(data= data1, x='X', y='Y')
```

```python
model = LinearRegression().fit(data1[['X']], data1[['Y']])
print(f"Intercept (c) : {model.intercept_[0]:.2f}")
print(f"Coefficient (m) : {model.coef_[0] [0]:.2f}")
```

```python
predicted_house_price = model.predict([[1800]])
print(predicted_house_price)
```

```python
y_all_predicted = model.predict(data1[['X']])
sns.scatterplot(data = data1, x='X', y='Y')
sns.lineplot(x=data1['X'], y= y_all_predicted.flatten())
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
r2 = r2_score(data1['Y'],y_all_predicted)
mse = mean_absolute_error(data1['Y'],y_all_predicted)
print(f" r2 score is {r2}")
print(f"MSE score is {mse}")
```
