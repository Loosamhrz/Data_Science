```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.DataFrame({
    'Temperature': [1,2,3,4,5,6,7,8],
    'Energy_Consumption': [2,5,9,15,14,10,5,1]
})
```

```python
sns.scatterplot(data = df, x="Temperature", y="Energy_Consumption")
```

```python
x= df[["Temperature"]]
y = df[["Energy_Consumption"]]
x_poly = PolynomialFeatures(degree=2).fit_transform(x)

model = LinearRegression().fit(x_poly,y)

energy_consumption_predicted = model.predict(x_poly)
print(f"energy consumption predicted {energy_consumption_predicted}")
```

```python
sns.scatterplot(data = df, x='Temperature', y = 'Energy_Consumption')
plt.plot(x,energy_consumption_predicted, color='red')
```

```python
x_poly = PolynomialFeatures(degree=2).fit_transform(x)
print(x_poly)
```

```python
energy_consumption_predicted = model.predict(PolynomialFeatures(degree=2).fit_transform([[2]]))
print(f"energy consumption predicted {energy_consumption_predicted}")
```

```python
from sklearn.metrics import r2_score
energy_consumption_predicted_all = model.predict(x_poly)
r2 = r2_score(df['Energy_Consumption'], energy_consumption_predicted_all)
print(f"{r2}")
```
