```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
prices = pd.Series([10,12,15,14,13,50],name='Price')
prices
```

```python
data_2d = prices.values.reshape(-1,1)
print(data_2d)
```

```python
#initialize Scalers
std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()

#Create a comparision dataframe
results = pd.DataFrame({
    'Original': prices,
    'Standardized(Z-Score)': std_scaler.fit_transform(data_2d).flatten(),
    'Normalized (0,1)': minmax_scaler.fit_transform(data_2d).flatten(),
    'Robust (IQR)': robust_scaler.fit_transform(data_2d).flatten()
})
print(results.round(2))
```
