```python
import pandas as pd
df = pd.read_csv('insurance.csv')
df
```

```python
df["sex"] = df["sex"].map({"male":1,"female":0})
df
```

```python
df["smoker"] = df["smoker"].map({"no":0,"yes":1})
df
```

```python
df.rename(columns={'sex':'gender'},inplace = True)
```

```python
df
```

```python
female_smoker=(df['gender']==0) & (df['smoker']==1)
print(female_smoker.sum())
```

```python
print(df.groupby('smoker')['charges'].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
smokers = df[df['smoker'] == 1]
sns.countplot(x='region', hue='gender', data=smokers, palette=['pink', 'skyblue'])
# Add nice labels
plt.title("Male vs Female Smokers by Region")
plt.xlabel("Region")
plt.ylabel("Count of Smokers")
plt.show()
```
